## Part 6 · Context 진행 보고 (주어짐)
오래 걸리는 작업은 `ctx: Context`로 진행률을 보고합니다. 클라이언트는 `progress_handler`로 받습니다 (실습 Lab 4).

````python
@mcp.tool
async def reindex(ctx: Context) -> dict:
    """문서를 다시 색인하며 진행률을 보고한다 (오래 걸리는 작업 예시)"""
    for i, d in enumerate(DOCS):
        await ctx.report_progress(i + 1, len(DOCS))
        await asyncio.sleep(0)          # 실제로는 임베딩/DB I/O
    await ctx.info("색인 완료")
    return {"reindexed": len(DOCS)}

async def on_progress(progress, total, message=None):
    print(f"  진행 {progress}/{total}")

async with Client(mcp, progress_handler=on_progress) as c:
    print((await c.call_tool("reindex", {})).data)
````

## Part 7 · 에이전트 시나리오
에이전트처럼 써 봅니다: **검색으로 근거 찾기 → 행동(메모) → 잘못된 입력 처리 → 상태 확인**.

````python
async with Client(mcp) as client:
    print("도구:", [t.name for t in await client.list_tools()])
    hits = (await client.call_tool("search_docs", {"query": "연차 신청 며칠 전", "k": 2})).data
    print("검색 근거:", [h["doc_id"] for h in hits])
    print("메모 저장:", (await client.call_tool("add_note", {"text": "연차 3영업일 전 신청"})).data)
    try:
        await client.call_tool("flag_doc", {"doc_id": "leave", "level": "초긴급"})
    except Exception as e:
        print("에러 전달:", str(e).splitlines()[-1])
    print("상태:", (await client.call_tool("stats", {})).data)
````

## Part 8 · 자동 채점 — 기능 점수
5개 항목으로 채점합니다. 5/5가 목표입니다.

````python
async def grade():
    score, log = 0, []
    async with Client(mcp) as c:
        tools = {t.name: t for t in await c.list_tools()}
        try:
            r = (await c.call_tool("search_docs", {"query": "연차", "k": 1})).data
            ok = bool(r) and any(x.get("doc_id") == "leave" for x in r)
        except Exception: ok = False
        score += ok; log.append(f"[{'O' if ok else 'X'}] 검색 도구 search_docs")

        try:
            await c.call_tool("add_note", {"text": "채점용"}); ok = "add_note" in tools
        except Exception: ok = False
        score += ok; log.append(f"[{'O' if ok else 'X'}] 부작용 도구 add_note")

        try:
            await c.call_tool("flag_doc", {"doc_id": "leave", "level": "없는값"}); ok = False
        except Exception: ok = True          # 잘못된 값에 에러를 던져야 정답
        score += ok; log.append(f"[{'O' if ok else 'X'}] 입력 검증 flag_doc")

        try:
            s = (await c.call_tool("stats", {})).data
            ok = getattr(s, "docs", None) == len(DOCS) or (isinstance(s, dict) and s.get("docs") == len(DOCS))
        except Exception: ok = False
        score += ok; log.append(f"[{'O' if ok else 'X'}] 구조화 출력 stats")

        ok = "stats" in tools and bool(getattr(tools["stats"].annotations, "readOnlyHint", False))
        score += ok; log.append(f"[{'O' if ok else 'X'}] annotations(readOnlyHint)")
    print("\n".join(log)); print(f"\n기능 점수: {score}/5")

await grade()
````
